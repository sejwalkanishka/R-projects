```r
# load the packages we need
library(tidyverse)
library(ggplot2)
library(skimr)
library(scales)
```

```r
# I noticed that there are 2 folders here, and they covers different time periods.
# After previewing the data, I found out that the date format is MM/DD/YYYY
# In this project, I will work on the data set 4.12.16 - 5.12.16
# "dailyActivity_merged" has already included all the daily data, so I decided to simply merge it with "sleepDay"
# I also considered merging "weightLogInfo", but the data is not measured daily

Daily_Activity <- read.csv("/kaggle/input/fitbit/mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/dailyActivity_merged.csv")
Sleep_Day <- read.csv("/kaggle/input/fitbit/mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/sleepDay_merged.csv")
```

```r
# Before merging the data into 1 table, let's check out if the primary keys (Id and ActivityDate) share the same format
head(Daily_Activity)
head(Sleep_Day)
```

# Pharse 3: Process (Data Cleaning)

```r
# There are 2 observations here
    # 1. The date format in these files are inconsistent, and all of them are <chr> not <date>

# To merge the dataframes successfully, we have to convert the format to "date"

Daily_Activity$ActivityDate <- as.Date(Daily_Activity$ActivityDate, format = "%m/%d/%Y")
Sleep_Day$SleepDay <- as.Date(Sleep_Day$SleepDay, format = "%m/%d/%Y %H:%M:%S")


# Review the data frame again
str(Daily_Activity$ActivityDate)
str(Sleep_Day$SleepDay)

# Note that the new date format is yyyy-mm-dd
```

```r
# Next, we need to rename the date column for all of them so that we can merge them

Daily_Activity <- rename(Daily_Activity, Date = ActivityDate)
Sleep_Day <- rename(Sleep_Day, Date = SleepDay)
```

```r
# Now, merge the date
# I originally tried "Merged <- merge(Daily_Activity, Daily_Calories, by = c("Id","Date"), all = TRUE)", but I found out that calories has been included in "Daily_Activity" already, so no need to merge

Merged <- merge(Daily_Activity, Sleep_Day, by = c("Id","Date"), all = TRUE)

head(Merged)

# We successfully merged all data selected now
# As mentioned before "Sleep_Day" was not measured daily, so there are a lot of N/A values

```

# Phrase 4: Analyse

```r
# Check what data we have now
colnames(Merged)
summary(Merged)
```

```r
# First, let's see if the steps data is stable overtime
ggplot(data=Merged, aes(x=Date, y=TotalSteps, color = TotalSteps))+
geom_smooth()+
geom_point()+
labs(title = "Total Steps/Date")
    
```

Ok, the total steps for the participants overtime is stable, that increases the reliability of the data set

Seems there is no clear correlation between them. I thought that the more steps we walk, the healthier (more sleep) we would be.

```r
# Let's check the correlation between calories and sleep time
ggplot(data=Merged, aes(x=Calories, y=TotalMinutesAsleep, color = TotalMinutesAsleep ))+
geom_smooth()+
geom_point()
```

The data points concentrate between 1500 to 3500 calories burnt, with a small peak at 2500 calories burnt, but the graph doesn't show a strong relationship between them.

## Next Step

Because there is no clear trend I can find from the raw data, I am going to categorise them into meaningful groups

### Calories

After doing the research about how many calories we should burn everyday.
I found out that:


*[How Many Calories Do I Burn in a Day?](https://www.webmd.com/fitness-exercise/how-many-calories-do-i-burn-in-a-day)*

Because we don't have the data about the sex and age of the participants, so I just took the average of 20 - 50 men and wemon.


And if we want to maintance our weight, we have to burn 300 active calories per day, and if we want to lose weight, the expected calories burned would be 500 - 700 calories. ([Amanda Roberts](https://www.autonomous.ai/ourblog/how-many-active-calories-per-day))

So I am going to categorise the data into 3 categories by calories burned:
1. Unhealthy: <2500 
2. Healthy: 2500 - 3200 (2500 + (300~700))
3. Highly Active: >3200

### Sleep time

According to the article by [Harvard Health Publishing](https://www.health.harvard.edu/blog/how-much-sleep-do-you-actually-need-202310302986), for most healthy adults, guidelines suggest at least 7 hours of sleep and sometimes [more than 9 hours of sleep is a sign of an underlying health problem](https://www.sleepfoundation.org/how-sleep-works/how-much-sleep-do-we-really-need).

I define the group as follows:
1. Under-Slept: < 420 min. (< 7 hrs)
2. Well-Rested: 420 - 540 min. (7 - 9 hrs)
3. Over-Slept: > 540 min. (> 9 hrs)


```r
# Categorise data by calories

Merged <- Merged %>%
    mutate(calories_group = case_when (
        Calories < 2500 ~ "Unhealthy",
        Calories >= 2500 & Calories <= 3200 ~ "Healthy",
        Calories > 3200 ~ "Highly Active" 
        )) %>%
    drop_na(calories_group)  # Remove rows with NA in TotalMinutesAsleep
```

```r
# Categorise data by sleep time

Merged <- Merged %>%
    mutate(sleep_group = case_when(
        TotalMinutesAsleep < 420 ~ "Under-Slept",
        TotalMinutesAsleep >= 420 & TotalMinutesAsleep <= 540 ~ "Well-Rested",
        TotalMinutesAsleep > 540 ~ "Over-Slept"
        )) %>%
    drop_na(sleep_group)

# Assign the order to the new groups
Merged$sleep_group <- factor(Merged$sleep_group, 
                            levels = c("Under-Slept", "Well-Rested", "Over-Slept"),
                            ordered = TRUE)
Merged$calories_group <- factor(Merged$calories_group, 
                               levels = c("Unhealthy", "Healthy", "Highly Active"),
                               ordered = TRUE)

# Check whether the code runs correctly or not
head(Merged)
```

```r
Merged %>%
    filter(calories_group == "Unhealthy")%>%
    ggplot()+
    geom_point(mapping = aes(x=Calories, y = TotalMinutesAsleep, color = sleep_group))+
    scale_x_continuous(limits = c(0, 5000)) +
      scale_y_continuous(limits = c(0, 800)) +
geom_smooth(aes(x=Calories, y = TotalMinutesAsleep))

Merged %>%
    filter(calories_group == "Healthy")%>%
    ggplot()+
    geom_point(mapping = aes(x=Calories, y = TotalMinutesAsleep, color = sleep_group))+
    scale_x_continuous(limits = c(0, 5000)) +
      scale_y_continuous(limits = c(0, 800)) +
geom_smooth(aes(x=Calories, y = TotalMinutesAsleep))

Merged %>%
    filter(calories_group == "Highly Active")%>%
    ggplot()+
    geom_point(mapping = aes(x=Calories, y = TotalMinutesAsleep, color = sleep_group))+
    scale_x_continuous(limits = c(0, 5000)) +
      scale_y_continuous(limits = c(0, 800)) +
geom_smooth(aes(x=Calories, y = TotalMinutesAsleep))
```

```r
Merged_percent <- Merged %>%
  group_by(calories_group, sleep_group) %>%
  summarise(count = n()) %>%
  group_by(calories_group) %>%
  mutate(perc = count / sum(count) * 100) %>%
  ungroup()

    
ggplot(Merged_percent, aes(x = calories_group, y = perc, fill = sleep_group)) +
  geom_bar(stat = "identity", position = "dodge")+
scale_y_continuous(limits = c(0, 100))+
labs(title = "Percentage of Sleep Groups by Calorie Group",
       x = "Calorie Group",
       y = "Percentage",
       fill = "Sleep Group")
```
